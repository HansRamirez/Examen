```python
# Inicialización de variables
pisos = 10
departamentos_por_piso = 4
tipos_departamento = ["A", "B", "C", "D"]
precios = {"A": 3800, "B": 3000, "C": 2800, "D": 3500}
departamentos_disponibles = [[True] * departamentos_por_piso for _ in range(pisos)]
compradores = {}


def mostrar_menu():
    print("")
    print("===== MENU =====")
    print("1. Comprar departamento")
    print("2. Mostrar departamentos disponibles")
    print("3. Ver listado de compradores")
    print("4. Mostrar ganancias totales")
    print("5. Salir")
    print("")

def comprar_departamento():
    print("===== Comprar departamento =====")
    print("")
    piso = int(input("Ingrese el número de piso (1-10): "))
    tipo = input("Ingrese el tipo de departamento (A, B, C, D): ").upper()

    # Validación de datos ingresados
    if piso < 1 or piso > pisos:
        print("¡Error! El número de piso es inválido.")
        print("")
        return
    if tipo not in tipos_departamento:
        print("¡Error! El tipo de departamento es inválido.")
        print("")
        return

    departamento = tipo + str(piso)

    # Verificar si el departamento está disponible
    if not departamentos_disponibles[piso - 1][tipos_departamento.index(tipo)]:
        print("¡Error! El departamento seleccionado no está disponible.")
        print("")
        return

    run = input("Ingrese el RUN del comprador (sin guion ni puntos): ")
    print("")

    # Registrar compra
    departamentos_disponibles[piso - 1][tipos_departamento.index(tipo)] = False
    compradores[run] = departamento

    print("La operación se ha realizado correctamente.")
    print("")

def mostrar_departamentos_disponibles():
    print("===== Departamentos Disponibles =====")
    print("")
    for piso in range(pisos):
        print(f"Piso {piso + 1}:")
        for tipo in tipos_departamento:
            departamento = tipo + str(piso + 1)
            estado = "Disponible" if departamentos_disponibles[piso][tipos_departamento.index(tipo)] else "Vendido"
            print(f"{departamento}: {estado}")
        print()


def ver_listado_compradores():
    print("===== Listado de Compradores =====")
    print("")
    for run in sorted(compradores):
        print(f"RUN: {run} - Departamento: {compradores[run]}")


def mostrar_ganancias_totales():
    print("===== Ganancias Totales =====")
    print("")
    total = sum(precios[tipo] for tipo in tipos_departamento)
    print(f"Ganancias totales: {total} UF")


# Función principal
def main():
    while True:
        mostrar_menu()
        opcion = input("Ingrese una opción (1-5): ")
        print("")

        # Validación de opción ingresada
        if opcion not in ["1", "2", "3", "4", "5"]:
            print("¡Error! Opción inválida.")
            print("")
            continue

        if opcion == "1":
            comprar_departamento()
        elif opcion == "2":
            mostrar_departamentos_disponibles()
        elif opcion == "3":
            ver_listado_compradores()
        elif opcion == "4":
            mostrar_ganancias_totales()
        else:
            # Opción 5: Salir del programa
            fecha_actual = datetime.date.today().strftime("%d/%m/%Y")
            print(f"Gracias por utilizar el sistema. Salida del programa el {fecha_actual}.")
            break


# Llamada a la función principal
main()


```
